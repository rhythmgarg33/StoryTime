```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies
```

```python
movies.iloc[0][3]
```

```python
import ast
```

```python
## text is basically a string
## run on each row for genre
## ast.literal_eval used to convert string into list of strings
##For us (human) it is very easy to tell that it is a dictionary data structure, but for Python, it is not easy, until it uses literal_eval function of ast class.
##We can use ast.literal_eval() here to evaluate the string as a python expression. It safely evaluates an expression node or a string containing a Python expression. The string or node provided may only consist of the following Python literal structures: strings, numbers, tuples, lists, dicts, booleans, and None.
##{'id': 28, 'name': 'Action'}
def extract(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(extract)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(extract)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
def fun(text):
    return text.split()
    
```

```python
## convert string to a list
movies['overview'] = movies['overview'].apply(fun)
movies.sample(5)
```

```python
movies
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
def fun1(text):
    return " ".join(text)
    
```

```python
## he join() method takes all items in an iterable and joins them into one string.
new['tags'] = new['tags'].apply(fun1)
new.head()
```

```python
def fun2(text):
    return text.lower()
```

```python
new['tags']=new['tags'].apply(fun2)
```

```python
new.head()
```

```python
# import these modules
## many words like dance,danced,dances----- to convert them to one common name.
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

ps = PorterStemmer()
```

```python
def stem1(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags']=new['tags'].apply(stem1)
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv
```

```python
## text vectorization
## covert a given text to vector
## technique used is bagofwords
## .toarray to convert it to numpy array
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
## to get most frequent words in the corpus
cv.get_feature_names()
```

```python
## high dimensions eucledian distance is not a good measure
## so we use cosine distance--- angle between two vectors
## distance inversely propotional to similarity
## when we find similarity-- later we sort to find top 10 movies
## but index position will be lost
## to avoid that we use enumerate
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
## distance of each movie with every other movie is calculated.
similarity = cosine_similarity(vector)
similarity.shape
```

```python
new[new['title']=='Avatar'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend("Batman Returns")
```
